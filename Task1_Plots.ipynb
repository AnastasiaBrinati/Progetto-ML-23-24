```python
import matplotlib.pyplot as plt

# Dati delle varie simulazioni
simulations = {
    'fedavg': [task1_fedavg_5, task1_fedavg_10, task1_fedavg_100],
    'qfedavg': [task1_qfedavg_5, task1_qfedavg_10, task1_qfedavg_100],
    'fedprox': [task1_fedprox_5, task1_fedprox_10, task1_fedprox_100]
}

metrics = ['Loss', 'Accuracy', 'Precision', 'Recall', 'F1 Score']
num_clients = ['5', '10', '100']

# Creazione dei grafici per ciascuna metrica
for metric in metrics:
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9,5))
    fig.suptitle(metric)

    # Iterazione sulle strategie
    for ax, (strategy, strategy_simulations) in zip(axes, simulations.items()):
        # Plot per ciascuna strategia
        for i, clients in enumerate(num_clients):
            label = f"{strategy}_{clients}"
            ax.plot(strategy_simulations[i]['Round'], strategy_simulations[i][metric], label=label)

        ax.set_xlabel('Round/Epoche')
        ax.set_ylabel(metric)
        ax.legend()
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```
