```python
import matplotlib.pyplot as plt

# Definition of metrics and strategies to be visualized
metrics = ['Loss', 'Accuracy', 'Precision', 'Recall', 'F1 Score']
strategies = ['task1_FedAvg', 'task1_QFedAvg', 'task1_FedProx']
num_clients_list = ['5', '10', '100']

# Creating subplots
fig, axs = plt.subplots(len(metrics), len(num_clients_list), figsize=(15, 20))

# Looping over each metric and number of clients to create subplots
for i, metric in enumerate(metrics):
    for j, num_clients in enumerate(num_clients_list):
        axs[i, j].set_title(f"{metric} - {num_clients} clients")
        axs[i, j].set_xlabel('Round' if 'Fed' in strategies[0] else 'Epoch')
        axs[i, j].set_ylabel(metric)

        # Plotting for the centralized result
        axs[i, j].plot(task1_centralizzato['Epoch'], task1_centralizzato[metric], label='Centralizzato', linestyle='--', color='black')

        # Plotting for each strategy
        for strategy in strategies:
            strategy_key = f"{strategy}_{num_clients}"
            axs[i, j].plot(eval(strategy_key)['Round' if 'Fed' in strategy else 'Epoch'], eval(strategy_key)[metric], label=strategy.split('_')[1], linestyle='-')

        # Adding the legend
        axs[i, j].legend()

# Optimizing the layout
plt.tight_layout()
plt.show()
```
