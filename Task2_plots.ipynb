```python
import matplotlib.pyplot as plt

# Definizione delle metriche e delle strategie da visualizzare
metrics = ['R2 Score', 'Mean Squared Error', 'Mean Absolute Percentage Error','Mean Squared Logarithmic Error','Loss']
strategies = ['task2_fedavg', 'task2_qfedavg', 'task2_fedprox']
num_clients_list = ['5', '10', '100']

# Creazione dei subplots
fig, axs = plt.subplots(len(metrics), len(num_clients_list), figsize=(15, 20))

# Looping su ogni metrica e numero di clienti per creare i subplots
for i, metric in enumerate(metrics):
    for j, num_clients in enumerate(num_clients_list):
        axs[i, j].set_title(f"{metric.replace('_', ' ').title()} - {num_clients} clients")
        axs[i, j].set_xlabel('Epoch')
        axs[i, j].set_ylabel(metric)

        # Plotting per ogni strategia per tutti gli altri casi
        for strategy in strategies:
            strategy_key = f"{strategy}_{num_clients}"
            if metric in eval(strategy_key):
                axs[i, j].plot(eval(strategy_key)['Epoch'], eval(strategy_key)[metric], label=strategy.split('_')[1], linestyle='-')

        # Aggiunta della metrica centralizzata
        central_metric = task2_centralizzato[metric]
        axs[i, j].plot(task2_centralizzato['Epoch'], central_metric, label="Centralizzato", linestyle='--', color='black')

        # Aggiunta della legenda
        axs[i, j].legend()

# Ottimizzazione del layout
plt.tight_layout()
plt.show()
```
